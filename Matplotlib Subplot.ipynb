```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
```

To avoid over lap between the figures

```python
fig,axes = plt.subplots(nrows=3,ncols=1)

axes[0].plot(x,y)

fig.tight_layout()
```

Manual way

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(a,b)

axes[0][1].plot(x,y)

axes[1][0].plot(a,b)

axes[1][1].plot(x,y)

fig.subplots_adjust(wspace=0.9, hspace =0.5 )
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(a,b)

axes[0][1].plot(x,y)
axes[0][1].set_title('Title 0 1')

axes[1][0].plot(a,b)
axes[1][0].set_ylabel('Y label 10')

axes[1][1].plot(x,y)

fig.suptitle('Figure Level',fontsize=16)
fig.tight_layout()
```

Saving Subplot

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(10,10),dpi=200)

axes[0][0].plot(a,b)

axes[0][1].plot(x,y)
axes[0][1].set_title('Title 0 1')

axes[1][0].plot(a,b)
axes[1][0].set_ylabel('Y label 10')

axes[1][1].plot(x,y)

fig.suptitle('Figure Level',fontsize=16)

fig.tight_layout()

fig.savefig('new_subplots.png',bbox_inches='tight')
```
