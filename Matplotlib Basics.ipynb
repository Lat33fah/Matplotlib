plt.show() is necesarry while working with .py scripts

```python
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("String Title");
```

```python
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.xlim(0,6)
plt.ylim(0,15)
plt.title("String Title");
```
