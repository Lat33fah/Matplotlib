```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(a,b)

plt.show()
```

Adding another set of axes to the figure

```python
fig = plt.figure()

# LAREGE AXES

axes1= fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_ylabel('A')
axes1.set_xlabel('B')
axes1.set_title('Power of 4')
# SMALL AXES

axes2= fig.add_axes([1,1,2,2])
axes2.plot(a,b)
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_ylabel('A')
axes2.set_xlabel('B')
axes2.set_title('Zoomed in')



plt.show()
```

```python
fig = plt.figure(dpi=100)

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)
```
