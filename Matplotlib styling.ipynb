```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,11)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x,x**2,label='X vs X^2')

ax.legend(loc = 'lower left')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x,x**2,label='X vs X^2')

ax.legend(loc =(1.2,0.6))
```

Visual Styling

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='green')

```

Using the RGB HEX code

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#8648a1',label='X vs X')

ax.plot(x,x+1,color='yellow',label='X vs X+1')

ax.legend()
```

Line Width

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#8648a1',lw=10)

ax.plot(x,x+1,color='yellow',linewidth=2)

```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#8648a1',lw=10,ls='--')

ax.plot(x,x+1,color='yellow',linewidth=5,linestyle='-.')

```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#8648a1',lw=10,ls='--',marker='o',ms=10)

ax.plot(x,x+1,color='red',linewidth=1,linestyle='-.',marker='+',markersize=20)

```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])


ax.plot(x, x, color="black", lw=1, ls='-', marker='s', markersize=20, 
        markerfacecolor="red", markeredgewidth=8, markeredgecolor="blue");
```
